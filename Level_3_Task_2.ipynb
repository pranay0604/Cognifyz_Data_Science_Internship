- Analyze the relationship between the type of
cuisine and the restaurant's rating.

- Identify the most popular cuisines among
customers based on the number of votes.

- Determine if there are any specific cuisines
that tend to receive higher ratings.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/pranay/Downloads/Cognifyz DS Internship/Dataset.csv")
```

```python
# Split multiple cuisines into first cuisine for analysis

df["Main Cuisine"] = df["Cuisines"].astype(str).apply(lambda x: x.split(",")[0])
```

```python
# Avg rating by cuisine

avg_rating_cuisine = df.groupby("Main Cuisine")["Aggregate rating"].mean().sort_values(ascending=False)
print("Top 10 cuisines by average rating :\n", avg_rating_cuisine.head(10))
```

```python
# Most popular cuisines by votes

popular_cuisine_votes = df.groupby("Main Cuisine")["Votes"].sum().sort_values(ascending=False)
print("Top 10 cuisines by total votes :\n", popular_cuisine_votes.head(10))
```

```python
# Higher rated cuisines (rating > 4)

high_rated = avg_rating_cuisine[avg_rating_cuisine > 4.0]
print("Cuisines with avg rating > 4 :\n", high_rated.head(10))
```
