**Task 3 : Geospatial Analysis**

- Visualize the locations of restaurants on a
map using latitude and longitude
information.

- Analyze the distribution of restaurants
across different cities or countries.
Determine if there is any correlation
between the restaurant's location and its
rating.

```python
import folium
import pandas as pd

df = pd.read_csv("/Users/pranay/Downloads/Cognifyz Internship Task/Dataset.csv")
```

```python
# Make sure columns exist
if "Latitude" in df.columns and "Longitude" in df.columns:
    # Create base map using average location
    center_lat = df["Latitude"].mean()
    center_lon = df["Longitude"].mean()

    restaurant_map = folium.Map(location=[center_lat, center_lon], zoom_start=4)

    # Add markers
    sample_df = df.sample(200, random_state=42)

    for _, row in sample_df.iterrows():
        folium.CircleMarker(
            location=[row["Latitude"], row["Longitude"]],
            radius=3,
            popup=f"Rating: {row['Aggregate rating']}",
            fill=True
        ).add_to(restaurant_map)

    restaurant_map.save("restaurants_map.html")
    print("Map saved as restaurants_map.html")
```

```python
# Distribution by City/Country Code

print("\nRestaurants by City (Top 10):\n", df["City"].value_counts().head(10))
print("\nRestaurants by Country Code (Top 10):\n", df["Country Code"].value_counts().head(10))
```

```python
# Correlation between location and rating

corr_lat = df["Latitude"].corr(df["Aggregate rating"])
corr_lon = df["Longitude"].corr(df["Aggregate rating"])
print("\nCorrelation Latitude vs Rating:", corr_lat)
print("Correlation Longitude vs Rating:", corr_lon)
```
