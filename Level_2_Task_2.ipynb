**Task 2 : Price Range Analysis**

- Determine the most common price range
among all the restaurants.

- Calculate the average rating for each price
range.

- Identify the color that represents the highest
average rating among different price ranges.

```python
import pandas as pd

df = pd.read_csv("/Users/pranay/Downloads/Cognifyz DS Internship/Dataset.csv")
```

```python
# Most common price range

most_common_price = df["Price range"].mode()[0]
print("Most common price range :", most_common_price)
```

```python
# Average rating by price range

avg_rating_price = df.groupby("Price range")["Aggregate rating"].mean().sort_values(ascending=False)
print("Avg Rating by Price Range :\n", avg_rating_price)
```

```python
# Color representing highest average rating

if "Rating color" in df.columns:
    color_rating = df.groupby("Rating color")["Aggregate rating"].mean().sort_values(ascending=False)
    print("Avg rating by Rating Color :\n", color_rating)
    print("Highest avg rating color :", color_rating.index[0])
```
